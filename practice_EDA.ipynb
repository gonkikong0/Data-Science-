# Query() Command 
To find/figure out what might be the reason for duplicate columns. <br>
Query() - Query the columns of a DataFrame with a boolean expression.

```python
#Checking an example duplicate
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
# ~ means inverse of the condition
df.reset_index(drop= True)
df = df.loc[~df.duplicated(subset= ['Coaster_Name', 'Location', 'Opening_Date_Clean'])].copy()
```
