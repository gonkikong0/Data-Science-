# Cleaning Phone Numbers 

Goal is to convert all of them in an uniform format (123-123-123). <br>
Function to be used - pandas.Series.str.replace(pattern, replacement) <br>

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]',"", regex=True) 
df
```

```python
# REGEX (Regular Expressions) are a mini-language used for searching, matching, 
# and replacing text patterns.
# 
# In older versions of pandas, str.replace(pattern, replacement) automatically 
# treated 'pattern' as a Regular Expression.
# 
# However, in newer versions (v1.4+), pandas treats 'pattern' as a literal string 
# by default. To use a Regular Expression, we now need to explicitly specify 
# regex=True.
#
# Example:
# df["col"].str.replace('[^a-zA-Z0-9]', '', regex=True)
# â†’ Removes all characters except letters and numbers.
```

# Splitting Columns
Now, looking at the Address Column. Suppose the employees want the Address column to be split into 3 part(s), making it cleaner and easier to read and comprehend. <br>
The data here, is separated by "," therefore we can separate the data using function split()

```python
df[['Street_Address', 'State', 'Zip_Code']] = df["Address"].str.split(',', n=2, expand = True)
df
```

# Understanding Code and Parameters -

Pandas.String.split() = It is used to split strings around given separator/delimeter <br>
[ A <b>Delimeter</b> is a character or a sequence of characters used to separate distinct elements in the Data. ] <br>
### Parameters - 

split(pat) = pat is the separator/delimeter (',' or '_') <br>
<b>n</b> = Number of splits in an output. <br>
<b>expand</b> = True/False | False by default | True returns the data split in multi-dimensionality, False returns in a form of split list.

# Standardizing Column Values using replace()
Function : <b>replace</b> <br>}
Columns : Do_Not_Contact , Paying Customer

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('No', 'N')
```

```python
df = df.fillna('')
df
```

# Fill NA
Pandas.DataFrame.fillna() = Fills <b>NA/NaN</b> values using the specified method. <br>
The data is individually cleaned, <i>but</i> the call centre only wants the information of those who can be called.

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Y":
        df.drop(x, inplace = True)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
df
#Another way
#df.dropna(subset = "Phone_Number"), inplace = True)
```
