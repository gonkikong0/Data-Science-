# Removing Duplicates
function : pandas.DataFrame.drop_duplicates() - Return DataFrame with duplicate rows <b>removed</b>

```python
df = df.drop_duplicates()
```

# Dropping useless, unnecessary columns 

function : pandas.DataFrame.drop(labels , columns)<br>
For example - Column "Not_Useful_Column" can be dropped. Hence 

```python
df.columns
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

# Removing out extra characters -

To remove extra characters like ... or / or _ in the column <b>Last_Name</b>.<br>
One way to do is to use Pandas.DataFrame.str.strip() function.

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

# Cleaning Phone Numbers 

Goal is to convert all of them in an uniform format (123-123-123). <br>
Function to be used - pandas.Series.str.replace(pattern, replacement) <br>

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]',"", regex=True) 
df
```
